```python
import pandas as pd
low_memory=False
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p_2020 = pd.read_csv('perfil_eleitor_secao_2020_SP.csv', delimiter=';', encoding ='latin1', error_bad_lines=False,
                     usecols = ['NM_MUNICIPIO', 'NR_ZONA', 'DS_GENERO','DS_FAIXA_ETARIA', 'DS_GRAU_ESCOLARIDADE'])
```

```python
p_2020.info()
```

```python
p_2020.shape
```

## Eleitores de São José dos Campos por gênero declarado



```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['DS_GENERO'].value_counts().plot(kind='bar', color = ['red', 'blue'])
```

## Eleitores de São José dos Campos por zona eleitoral


```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['NR_ZONA'].value_counts().plot(kind='bar')
```

## Eleitores de São José dos Campos por grau de escolaridade

```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['DS_GRAU_ESCOLARIDADE'].value_counts().plot(kind='bar')
```

## Eleitores de São José dos Campos faixa etária

```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['DS_FAIXA_ETARIA'].value_counts().plot(kind='barh', figsize = (10,10))
```
