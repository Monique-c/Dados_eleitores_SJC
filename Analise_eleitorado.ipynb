```python
import pandas as pd
low_memory=False
import matplotlib.pyplot as plt
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
p_2020 = pd.read_csv('perfil_eleitorado_2020.csv', delimiter=';', encoding ='latin1', error_bad_lines=False,
                     usecols = ['NM_MUNICIPIO', 'NR_ZONA', 'DS_GENERO','DS_FAIXA_ETARIA', 'DS_GRAU_ESCOLARIDADE', 'QT_ELEITORES_PERFIL'])
```

```python
p_2020.info()
```

```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['DS_GENERO'].value_counts().plot(kind='barh', figsize = (10,6), fontsize = 15, color = ['red', 'blue']).set_title('ELEITORES POR GÊNERO', fontsize = 21)
```

```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['NR_ZONA'].value_counts().plot(kind='bar',figsize = (6,6), fontsize = 15).set_title('ELEITORES POR ZONA ELEITORAL', fontsize = 21)
```

```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['DS_GRAU_ESCOLARIDADE'].value_counts().plot(kind='barh', figsize = (10,10), fontsize = 15).set_title('ELEITORES POR GRAU DE ESCOLARIDADE', fontsize = 21)
```

```python
sjc = p_2020.query('NM_MUNICIPIO == "SÃO JOSÉ DOS CAMPOS"')['DS_FAIXA_ETARIA'].value_counts().plot(kind='barh', fontsize = 15, figsize = (10,10)).set_title('ELEITORES POR FAIXA ETÁRIA', fontsize = 21)
```
